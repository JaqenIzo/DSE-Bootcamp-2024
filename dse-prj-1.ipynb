**7. What is the mean value of each column for each weather condition?**

```python
means=[]
conditions = df['Weather_Condition'].unique()
numerical_cols = [i for i in df.columns if df[i].dtype in ['int64','float64']]
for cond in conditions:
    mean_cond = df[df['Weather_Condition'] == cond].loc[:,numerical_cols].mean().values
    means.append((cond,mean_cond))
        
means

df_means = pd.DataFrame(data=[i[1] for i in means],columns= numerical_cols,index=conditions)
df_means.head()
    
```
